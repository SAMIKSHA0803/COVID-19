```python
#importing covid vaccine dataset

vaccine_df = pd.read_csv("C:/Users/hp/Desktop/python covid/covid_vaccine_statewise.csv")
```

```python
# checking first 10 rows of the data set

vaccine_df.head(10)
```

```python
#dropping unnecessary columns from covid_19 dataset

covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df.head()
```

```python
#Changing the date format

covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
covid_df.head()
```

```python
# finding active no of cases: confirmed cases - cured cases - deaths

covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
# creating a new variable - STATEWISE , AND PIVOT TABLE
statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
```

```python
#creating a new variable as - Recovery rate
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
#creating a new variable as: Mortality rate
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# top 10 states based on the active no of cases

top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by = ['Active_cases', 'Date'], ascending = False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 States With most active case in India", size =25)
```

```python
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'red')
```

```python
#final plot of top 10 states with no of active cases

top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by = ['Active_cases', 'Date'], ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States With most active case in India", size =25)
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'red')
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#top 10 states with highest deaths

top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
fig = plt.figure(figsize=(18,5))

plt.title("Top 10 states with most Deaths", size = 25)

ax = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths", x = "State/UnionTerritory", linewidth =2, edgecolor = 'black')

plt.xlabel("States")
plt.ylabel("Total Death Cases")
```

```python
#line plot to see growth or trend of active cases of top 5 states

fig = plt.figure(figsize = (12, 6))

ax = sns.lineplot(data = covid_df[covid_df['States/UnionTerritories'].isin(['Maharashtra', 'Karnatka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh']), x = 'Date', y = 'Active_cases', hue = 'States/UnionTerritory'])

ax.set_title("Top 5 Affected States in India", size = 16 )
```

```python
#dropping unnecessary columns

vaccination = vaccine_df.drop(columns = ['Sputnik V (Doses Administered)', 'AEFI', '18-44 Years(Individuals Vaccinated)', '45-60 Years(Individuals Vaccinated)', '60+ Years (Doses Administered)'], axis = 1)

vaccination.head()
```

```python
# drop the states where state is India

vaccine = vaccine_df[vaccine_df.State != 'India']
vaccine.head(10)
```

```python
#renaming one of the columns

vaccine.rename(columns = {"Total Individuals Vaccinated" : "Total"}, inplace = True)
vaccine.head(10)
```

```python
# most vaccinated states

max_vac = vaccine.groupby('State')['Tota'].sum().to_frame('Tota')
max_vac = max_vac.sort_values('Tota', ascending = False)[:5]

max_vac
```

```python
#converting into chart

fig = plt.figure(figsize = (10,5))
plt.title("Top 5 Vaccinated States in India", size = 20)
x = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Tota, x = max_vac.index, linewidth = 2, edgecolor = 'Black')
plt.xlabel("States")
plt.ylabel("vaccination")
plt.show()
```

```python
#Least vaccinated states

least_vac = vaccine.groupby('State')['Tota'].sum().to_frame('Tota')
least_vac = least_vac.sort_values('Tota', ascending = True )[:5]

least_vac
```

```python
#converting it into a chart

fig = plt.figure(figsize = (10,5))
plt.title("Least 5 Vaccinated States in India", size = 20)
x = sns.barplot(data = least_vac.iloc[:10], y = least_vac.Tota, x = least_vac.index, linewidth = 2, edgecolor = 'Black')
plt.xlabel("States")
plt.ylabel("vaccination")
plt.show()
```
